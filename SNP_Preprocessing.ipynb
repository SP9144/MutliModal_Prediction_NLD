```python
import os
import glob
import allel
import numpy as np
import pandas as pd
```

```python
os.chdir("./SNPs")
```

```python
snp_ids = dict()

for file in glob.glob("*.vcf"):
    df = allel.vcf_to_dataframe(file)
    rows = len(df)
    for i in range(rows):
        key_id = df['CHROM'][i] + '_' + str(df['POS'][i])+ '_' + df['REF'][i] + '_' + df['ALT_1'][i]
#         print(key_id)
        if key_id in snp_ids:
            snp_ids[key_id] += 1
        else:
            snp_ids[key_id] = 1
```

```python
print("Unique SNPs =", len(snp_ids))
```

```python
cols = ["PatNo"] + list(snp_ids.keys())
snp_df = pd.DataFrame(0, columns=cols, index=range(733))

row = 0
for file in glob.glob("*.vcf"):
    patNo = file[0:4]
    print(row, "-", patNo)
    snp_df.loc[row]["PatNo"] = patNo
    df = allel.vcf_to_dataframe(file)
    rows = len(df)
    for i in range(rows):
        col = df['CHROM'][i] + '_' + str(df['POS'][i])+ '_' + df['REF'][i] + '_' + df['ALT_1'][i]
        snp_df.loc[row][col] = 1
    row += 1
```
